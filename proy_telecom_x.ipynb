#Enlace Datos Json desde Github Telecom-x



```python
import requests
import pandas as pd
url ="https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)
df
```

# NormalizaciÃ³n de Datos


```python
import pandas as pd
df_planta = pd.json_normalize(data)
df_planta
```

```python
df_planta.info()
```

#Revision de Valores Nulos

```python
df_planta = df_planta[df_planta['Churn'].str.strip() != '']
print("Cantidad de Nulos", df_planta.isnull().sum())
```

#Cantidad de vacios

```python
print("Numero de Nulos", df_planta.isnull().sum())
```

#Cambiar tipos de datos

```python
df_planta['account.Charges.Total'] = pd.to_numeric(df_planta['account.Charges.Total'], errors='coerce')
print(df_planta['account.Charges.Total'].dtype)
```

```python
df_planta = df_planta[df_planta['Churn'].str.strip() != '']
print("Numero de filas totales':", len(df_planta))
```

```python
df_planta.describe()
```

#Cantidad de cuentas diarias

```python
df_planta.loc['Cuentas_Diarias'] = df_planta['account.Charges.Monthly'] / 30
```

```python
print(df_planta.columns)
```

```python
df_planta['Cuentas_Diarias'] = df_planta['account.Charges.Monthly'] / 30
print(df_planta[['account.Charges.Monthly', 'Cuentas_Diarias']])
```

##Sexo Biologico de los Cientes

```python
import plotly.express as px
fig1 = px.histogram(df_planta, x = 'customer.SeniorCitizen', text_auto = True, color = 'Churn', barmode = 'group',title='Histograma de Evasion por Sexo', color_discrete_sequence=px.colors.qualitative.Pastel,
                    labels={'Churn' :'Evasion'})

```
